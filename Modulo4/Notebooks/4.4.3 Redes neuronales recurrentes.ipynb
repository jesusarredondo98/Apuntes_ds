#### Train Test Split

Nota: Se aplica la metodología habitual de *split* de series de tiempo.

```python
len(df) #Validamos dimensiones
```

```python
val_percent = 0.1 #Partición de entrenamiento, no podemos hacer para series de tiempo los valores aleatorios
```

```python
len(df) * val_percent
```

```python
val_point = np.round(len(df)*val_percent)
```

```python
val_ind = int(len(df) - val_point) #construimos la particion
```

```python
val_ind
```

```python
train = df.iloc[:val_ind]
val = df.iloc[val_ind:]
```

#### Preprocesamiento

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler() #Instanciamos el rescalado
```

```python
scaler.fit(train)
```

```python
Xs_train = scaler.transform(train)
Xs_val = scaler.transform(val)
```

```python
true_val = scaler.inverse_transform(Xs_val)
#Al trabajar con series de tiempo vale la pena usar la transformacion inversa en nuestra porcion de validación para tener la variable original
#Para las predicciones tener el valor real
```

#### Time Series Generator

Esta clase toma una secuencia de observaciones recopilados en
intervalos iguales, junto con parámetros de series de tiempo como
stride, longitud de la historia, etc., para producir lotes para las porciones de
training/validation.

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# help(TimeseriesGenerator)
```

```python
# Definición del generador

length = 2       # Longitud de las secuencias de salida (en número de pasos en el tiempo)
batch_size = 1   # Número de muestras de series de tiempo en cada uno de los lotes
generator = TimeseriesGenerator(data=Xs_train, targets=Xs_train, length=length, batch_size=batch_size)
```

### Modelación

```python
length = 49        # Longitud de las secuencias de salida (en número de pasos de tiempo)
batch_size = 1     # Número de muestras de series de tiempo en cada lote
```

```python
generator = TimeseriesGenerator(Xs_train, Xs_train, length=length, batch_size=batch_size)
validation_generator = TimeseriesGenerator(Xs_val, Xs_val, length=length, batch_size=batch_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=2) #Hay que hacer un diagnostico previo
```

#### RNN

```python
from tensorflow.keras.layers import SimpleRNN #Red Neuronal Recurrente simple
```

```python
# Se utiliza solamente una variable en nuestra serie temporal
n_features = 1 #Al ser univariada es igual a 1
```

```python
# Definición del modelo
model_rnn = Sequential()

# Capa oculta única de la RNN
model_rnn.add(SimpleRNN(units=50, input_shape=(length,n_features))) #Numero de unidades, input shape 49,1

# Capa de salida
model_rnn.add(Dense(1)) #Capa densa con problema de regresion

model_rnn.compile(optimizer='adam', loss='mse')
```

```python
model_rnn.summary() #Es un modelo pequeño
```

```python
%%time
model_rnn.fit_generator(generator, epochs=10, validation_data=validation_generator, callbacks=[early_stop])
#Con generadores es mas util usar el fit_generator
#No poner muchas epocas
```

```python
losses = pd.DataFrame(model_rnn.history.history)
losses.plot();
```

```python
length
```

```python
first_eval_batch = Xs_train[-length:] #Pronostico con el entrenamiento de tamaño 49
first_eval_batch
```

```python
first_eval_batch = first_eval_batch.reshape((1, length, n_features)) #Redimensionamos para poner el tamaño del batch
first_eval_batch.shape
```

```python
model_rnn.predict(first_eval_batch)
```

```python
Xs_val[0]
```

Se coloca ahora esta lógica dentro de un bucle `for` para predecir todo el rango de prueba.

```python
val_predictions = []

first_eval_batch = Xs_train[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(Xs_val)):
    
    # Obtiene la primera predicción por delante ([0] es para tomar solo el elemento en lugar del arreglo)
    current_pred = model_rnn.predict(current_batch)[0]
    
    # Almacena predicción
    val_predictions.append(current_pred) 
    
    # Actualiza el lote para incluir la predicción y eliminar el primer valor
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis=1)
```

```python
val_predictions
```

```python
Xs_val
```

```python
val.head()
```

```python
true_predictions = scaler.inverse_transform(val_predictions)
val['RNN Predictions'] = true_predictions
val.plot(figsize=(12,8));
```

```python
print(f"MSE: {mean_squared_error(y_true=true_val, y_pred=true_predictions)}")
print(f"MAE: {mean_absolute_error(y_true=true_val, y_pred=true_predictions)}")
print(f"R2: {r2_score(y_true=true_val, y_pred=true_predictions)}")
```

#### LSTM

```python
from tensorflow.keras.layers import LSTM
```

```python
# Definición del modelo
model_lstm = Sequential()

# Capa oculta única de la RNN
model_lstm.add(LSTM(units=50, input_shape=(length,n_features)))

# Capa de salida
model_lstm.add(Dense(1))

model_lstm.compile(optimizer='adam', loss='mse')
```

```python
true_predictions = scaler.inverse_transform(val_predictions)
val['LSTM Predictions'] = true_predictions
val.plot(figsize=(12,8));
```

```python
print(f"MSE: {mean_squared_error(y_true=true_val, y_pred=true_predictions)}")
print(f"MAE: {mean_absolute_error(y_true=true_val, y_pred=true_predictions)}")
print(f"R2: {r2_score(y_true=true_val, y_pred=true_predictions)}")
```

### Forecasting
