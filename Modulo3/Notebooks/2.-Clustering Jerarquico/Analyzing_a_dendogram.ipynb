```python
from sklearn.datasets import make_blobs
```

```python
from matplotlib.axes._axes import _log as matplotlib_axes_logger
matplotlib_axes_logger.setLevel('ERROR')
```

```python
nb_samples = 12
nb_centers = 4
```

```python
X, Y = make_blobs(n_samples=nb_samples, n_features=2, center_box=[-1, 1],
centers=nb_centers, random_state=1000)
```

```python
Y #Son los grupos clase objetivo
```

```python
sns.scatterplot(X[:,0],X[:,1],hue=Y,palette='Set2')
# Estan en 2 dim entonces no hay qye realizar un algoritmo de disminucion
```

```python
from scipy.spatial.distance import pdist #Sirve para encontrar los dendogramas
from scipy.cluster.hierarchy import linkage
dm = pdist(X, metric='euclidean') #Calculo de distancia euclideana
Z = linkage(dm, method='ward') #Metodo en el que se fija, calcula las distancias en pares
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram
fig, ax = plt.subplots(figsize=(12, 8))
d = dendrogram(Z, show_leaf_counts=True, leaf_font_size=14, ax=ax)
ax.set_xlabel('Samples', fontsize=14)
ax.set_yticks(np.arange(0, 6, 0.25))
plt.show()

#Vemos las unicones
```
