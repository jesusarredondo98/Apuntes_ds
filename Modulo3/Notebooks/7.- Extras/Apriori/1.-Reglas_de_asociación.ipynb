```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('../data/Market_Basket_Optimisation.csv', header = None)
```

```python
dataset #Se tiene registro de tickets y que se consumio, con transacciones variadas
```

```python
dataset.apply(list,axis=1) #Creamos una lista
```

```python
transactions=list(dataset.apply(list,axis=1)) #Generamos una lista de listas
```

```python
transactions
```

```python
transactions_clean=[]
#Quitamos los nan
for lista in transactions:
    lista= [x for x in lista if str(x) != 'nan']
    transactions_clean.append(lista)
```

```python
transactions_clean #Lista de listas
```

```python
transactions_clean[1] #Lista primer registro
```

```python
from mlxtend.frequent_patterns import apriori #Importar la libreria mlxtend, algoritmo apriori
```

```python
from mlxtend.frequent_patterns import association_rules #reglas de asociacion
```

```python
from mlxtend.preprocessing import TransactionEncoder #Transaction encoder
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(transactions_clean).transform(transactions_clean) #Entrenamos con la tranformacion
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df #Se ve como una vectorizacion
```

```python
apriori(df, min_support = 0.003) #Minimo de suporte, no queremos que sea muy bajo, no hay valor indicado
#Crea el soporte y los productos que lo tienen
```

```python
frequent_itemsets=apriori(df, min_support=0.03, use_colnames=True) #Cambiamos el minimo de soporte
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift') #No interesa mas el lift
```

```python
rules #Conjunto de reglas
```

```python
rules.sort_values('lift') #Todo lift mayor a 1 es util 
```
