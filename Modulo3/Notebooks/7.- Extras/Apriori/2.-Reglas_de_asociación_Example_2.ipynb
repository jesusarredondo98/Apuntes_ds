```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel('../data/Online Retail.xlsx',engine='openpyxl')
data.head() #Datos con una de retail online
```

```python
data.shape #Tama√±o
```

```python
data.isnull().sum() 
```

```python
data.dropna(inplace=True) #Wuitamos nulos
```

```python
dict_products=data[['StockCode','Description']].drop_duplicates().set_index('StockCode')['Description'].to_dict() #Auxiliar sin duplicados
```

```python
data.reset_index(drop=True,inplace=True) #Reiniciamos el index
```

```python
data['StockCode']= data['StockCode'].astype(str) #Pasamos el dato a texto
```

```python
customers = data["CustomerID"].unique().tolist() #Ponemos clientes
```

```python
import random
```

```python
random.shuffle(customers) #Creamos una muestra de los clientes
```

```python
customers_train = [customers[i] for i in range(round(0.3*len(customers)))]
```

```python
train_df = data[data['CustomerID'].isin(customers_train)] #Creacion del subconjunto
```

```python
purchases_train = []

for i in customers_train:
    temp = train_df[train_df["CustomerID"] == i]["StockCode"].tolist()
    purchases_train.append(temp)
```

```python
purchases_train #lista de listas
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(purchases_train).transform(purchases_train)
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head() #Se tienen mas productos
```

```python
apriori(df, min_support = 0.03) #Identificamos un minimo de support
```

```python
frequent_itemsets=apriori(df, min_support=0.03, use_colnames=True)
```

```python
frequent_itemsets.head()
```

```python
rules=association_rules(frequent_itemsets,metric='lift')
```

```python
rules
```

```python
rules.sort_values('lift',ignore_index=True,inplace=True) #Ordenamos
```

```python
rules.head() #Mayores a 1
```

```python
rules[['name_antecedent','name_consequent']]=rules[['antecedents','consequents']].applymap(lambda x: tuple(map(dict_products.get,x))) #Agregamos los nombres
```
