```python

customers_train = [customers[i] for i in range(round(0.9*len(customers)))]


train_df = df[df['CustomerID'].isin(customers_train)]
validation_df = df[~df['CustomerID'].isin(customers_train)]
```

```python

purchases_train = []

for i in customers_train:
    temp = train_df[train_df["CustomerID"] == i]["StockCode"].tolist()
    purchases_train.append(temp)
```

```python
purchases_val = []


for i in validation_df['CustomerID'].unique():
    temp = validation_df[validation_df["CustomerID"] == i]["StockCode"].tolist()
    purchases_val.append(temp)
    
#Guardamos con animos de probar
```

```python
model = Word2Vec(window = 10, sg = 1, hs = 0,
                 negative = 10,
                 alpha=0.03, min_alpha=0.0007,
                 seed = 14)
#Vectorizacion
#ratio de aprendizage afecta la configuaracion de la red
#La ventana no debe de ser tan pequeña (mayor o igual a 5)
#Posiblemete se necesite hacer varias iteraciones
```

```python
model.build_vocab(purchases_train) #Pasamos productos y los identifica
```

```python
model.train(purchases_train, total_examples = model.corpus_count, 
            epochs=10, report_delay=1)
#Pasamos lo tiquets, corpues producto a ser vectorizados
#Las epocas son 
```

## Recomendaciones

```python
products = train_df[["StockCode", "Description"]]

products.drop_duplicates(inplace=True, subset='StockCode', keep="last")

products_dict = products.groupby('StockCode')['Description'].apply(list).to_dict()
```

```python
products_dict #Mapeamos el producto con su nombre
```

```python
#Medimos la similitud de producto con la similitud coseno
def similar_products(v, n = 6):
    
    ms = model.wv.similar_by_vector(v, topn= n+1)[1:]
    #n+1[:1] es un truco para que no llame al vector
    new_ms = []
    for j in ms:
        pair = (products_dict[j[0]][0], j[1])
        new_ms.append(pair)
        
    return new_ms     
```

```python
similar_products(model.wv['90019A'])
#Productos similares al producto
#sin el n+1[:1] traeria al mismo producto
```

```python
#Recomendación de varios productos que se consumen juntos
def aggregate_vectors(products):
    product_vec = []
    for i in products:
        try:
            product_vec.append(model.wv[i])
        except KeyError:
            continue
        
    return np.mean(product_vec, axis=0) #Se calcula la media de ambos vectores
#Calcular la media es peligrosa para productos alejados
```
