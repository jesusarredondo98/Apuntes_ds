```python
def pca_func(df, columns, pca_init = 3, alpha = .8):

    from sklearn.decomposition import PCA
    from sklearn.preprocessing import StandardScaler

    sc = StandardScaler()
    xsc = pd.DataFrame(sc.fit_transform(df[columns]), columns = columns, index = df.index)

    cumvar = 0

    while cumvar < alpha:
        pca = PCA(n_components = pca_init)
        xpca = pd.DataFrame(pca.fit_transform(xsc), columns = [f'cp_{i}' for i in range(pca_init)], index = df.index)
        cumvar = pca.explained_variance_ratio_.cumsum()[-1]
        pca_init +=1

    return(pca, xpca)
```

```python
red_dim = pca_func(df_estaciones, vcont, pca_init = 3)
```

## WoE

```python
def woe_estaciones(df,column,column_objetivo="Gender",step=10):
    aux_1 = df[[column_objetivo,column]].copy()
    if column in vcont:
        li = df[column].min()-5
        ls = df[column].max()+5
        step = step
        aux_1["column_cut"]=pd.cut(aux_1[column].astype(int), range(li,ls,step)).astype(str)
        aux_1.loc[aux_1["column_cut"]=="nan","colum_cut"]="No hay información"
        a =list(set(aux_1["column_cut"]))
        key = []
        for title in a:
            key.append(title.split(', ')[0][:-1])
        aux_dict=dict(zip(a,key))
        aux_1["column_categorica"]=aux_1["column_cut"].map(aux_dict)
    else:
        aux_1["column_cut"]=aux_1[column].astype(str)
        aux_1["column_categorica"]=aux_1["column_cut"]
    aux_2 = aux_1[['column_categorica', column_objetivo]].groupby('column_categorica').agg(['count', 'sum'])
    aux_2['evento_column'] = aux_2[column_objetivo, 'sum']
    aux_2['no_evento_column'] = aux_2[column_objetivo, 'count'] - aux_2['evento_column']
    aux_2['%evento_column'] = aux_2['evento_column']/aux_2['evento_column'].sum()
    aux_2['%no_evento_column'] = aux_2['no_evento_column']/aux_2['no_evento_column'].sum()
    aux_2['WoE_column'] = np.log(aux_2['%no_evento_column']/aux_2['%evento_column'])
    aux_2['IV'] = (aux_2['%no_evento_column'] - aux_2['%evento_column'])*aux_2['WoE_column']
    aux_2.loc[aux_2['IV'] == np.inf,'IV'] = 0
    iv = aux_2['IV'].sum()
    return aux_1,aux_2,iv
```
